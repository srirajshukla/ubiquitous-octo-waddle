```python
import requests
import pandas as pd

url = 'https://guidance.cdp.net/en/guidance?ctype=ExternalRef&idtype=RecordExternalRef&cid=C4.1a&otype=Guidance&incchild=0%C2%B5site=0&gettags=0'
html = requests.get(url).content
```

```python
html
```

```python
df_list = pd.read_html(html)
i=-1
for data in df_list:
    i+=1
    df = data
    print(df)
    df.to_csv(f'my{i}.csv')
    df
```
